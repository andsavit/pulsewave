```python
def reverb_chiama_rest(query = "", page = 1, per_page = 50):
    """Funzione di chiamata API REST a reverb. La query="" potrebbe ritornare tutti i listings"""

    load_dotenv()

    headers = {
        "Authorization": f"Bearer {os.getenv('REVERB_TOKEN')}",
        "Accept-Version": "3.0"
    }

    params = {
        "query": query,
        "page": page,
        "per_page": per_page,
        "sort": "published_at|desc"
    }

    response = requests.get("https://api.reverb.com/api/listings", headers=headers, params=params)

    if response.status_code == 200:
        return response.json()
        # salva/parsa come preferisci
    else:
        print(f"Errore: {response.status_code} - {response.text}")
        return None
```

## Caricamento dati su MongoDB

```python
client = MongoClient("mongodb://localhost:27017/")
db = client["strumenti_usato"]
col = db["reverb_storico"]

col.create_index("id", unique = True)
```

```python
#ottiene il numero di listings risultanti dalla query e lo usa come indice del for
header_listings = reverb_chiama_rest(per_page=1)
total_listings = header_listings.get("total", 0)
total_pages = (total_listings // 50) + 1

print(f"Totale annunci stimati: {total_listings} su {total_pages} pagine")

now = datetime.utcnow().isoformat()

resume = 409
last_page = None

#chiama le API una pagina alla volta e carica come batch
for i in tqdm(range(resume, total_pages + 1), desc="Scaricamento pagine"):
    ops = 0
    continua = False

    #esegue max 3 tentativi di chiamata per pagina
    while not continua and ops < 3:
        l_batch = reverb_chiama_rest(page=i)

        if l_batch and "listings" in l_batch: #verifica che la chimata abbia dato risultato valido
            #aggiunge campo tempo di aggiunta
            for l in l_batch["listings"]:
                l["uploaded_at"] = now

            try:
                col.insert_many(l_batch["listings"], ordered = False)
            except errors.BulkWriteError as e:
                duplicates = len(e.details['writeErrors'])
                print(f"{duplicates} duplicati ignorati.")

            continua = True
        else:
            print(f"Pagina {i} fallita al tentativo {ops+1}.")
            ops = ops+1
            time.sleep(1)
    last_page = i

print(f"Caricamento Terminato.")
```

```python
import os
import pandas as pd

def get_category_live_ids(product_type: str, category: str):
    file_path = "/Users/andrea/Progetti/strumenti_usato/data/reverb/used_IT/listings.parquet"
    if os.path.exists(file_path):
        df = pd.read_parquet(file_path)
        # Filter for this specific product_type/category and current records
        filtered = df[(df['product_type'] == product_type) & 
                     (df['category'] == category) & 
                     (df['snap_is_current'] == True)]
        return set(filtered['id'].tolist())
    return set()
```
